1. Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find the maximum score in the dictionary
max_score = max(employees.values())

# Identify all employees who have the maximum score
top_performers = [name for name, score in employees.items() if score == max_score]

# Format and display the output
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})")
```

2. Search Query Keyword Analysis

```python
query = input("Enter your query: ")

# Convert to lowercase and split into words (ignoring basic punctuation)
words = query.lower().split()

# Count the frequency of each word
word_counts = {}
for word in words:
    word_counts[word] = word_counts.get(word, 0) + 1

# Filter for keywords searched more than once
duplicates = {word: count for word, count in word_counts.items() if count > 1}

print(duplicates)
```

4. Email Domain Usage Analysis

```python
emails = ["ravi@gmail.com", "anita@yahoo.com", "kiran@gmail.com", "suresh@gmail.com", "meena@yahoo.com"]

domain_counts = {}
total_emails = len(emails)

# Extract domains and count occurrences
for email in emails:
    domain = email.split('@')[-1]
    domain_counts[domain] = domain_counts.get(domain, 0) + 1

# Calculate and display percentages
for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```
