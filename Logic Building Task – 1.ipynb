1. User Login Check

```python
username = "admin"
password = "1234"
entered_user = input ("Username: ")
entered_pass = input ("Password: ")
if entered_user == username and entered_pass == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

2. Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1
print(f"Total Pass: {pass_count}")
print(f"Total Fail: {fail_count}")
```

3. Simple Data Cleaner

```python
names = [" Alice", "bob", " CHARLIE "]
cleaned_names = [name.strip().lower() for name in names]
print("Cleaned names:",cleaned_names)
```

4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    length = len(msg)
    if length > 10:
        print(f"Message: '{msg}' | Length: {length} (Flag: Long Message)")
    else:
        print(f"Message: '{msg}' | Length: {length}")
```

5. Error Message Detector


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = logs.count("ERROR")
print(f"Total error count: {error_count}")
```
